```python
import pandas as pd
import numpy as np
import json
from tqdm.notebook import tqdm, trange
```

```python
credits_df = pd.read_csv("./archive/tmdb_5000_credits.csv")
credits_df.head(3)
```

## Cast

```python
cast_data = credits_df['cast'].map(lambda x: json.loads(x))

unique_cast = []
for i, cast in enumerate(tqdm(cast_data)):
    for cast_member in cast:
        if cast_member not in unique_cast:
            cast_member['movie_id'] = credits_df.iloc[i]['movie_id']
            unique_cast.append(cast_member)

cast_df = pd.json_normalize(unique_cast)
cast_df
```

```python
cast_members_df = cast_df.copy()
cast_members_df = cast_members_df.drop(columns=['cast_id', 'gender', 'name', 'order'])
cast_members_df.reset_index(inplace=True)
cast_members_df = cast_members_df.rename(columns={"id": "actor_id"})
cast_members_df = cast_members_df.rename(columns={"index": "id"})
cast_members_df = cast_members_df[['id', 'movie_id', 'actor_id', 'character', 'credit_id']]
cast_members_df['id'] = cast_members_df['id'].apply(lambda x: x + 1)
cast_members_df
```

## Actors

```python
actors_df = cast_df[['id', 'name', 'gender']].copy()
actors_df = actors_df.drop_duplicates(subset=['id'])
actors_df
```

## Crew

```python
crew_data_json = credits_df['crew'].map(lambda x: json.loads(x))
crew_data = crew_data_json.copy()

unique_crew = []
for i, crew in enumerate(tqdm(crew_data)):
    for crew_member in crew:
        if crew_member not in unique_crew:
            crew_member['movie_id'] = credits_df.iloc[i]['movie_id']
            unique_crew.append(crew_member)

crew_df = pd.json_normalize(unique_crew)
crew_df
```

```python
movie_crew_df = crew_df.copy()
movie_crew_df = movie_crew_df.drop(columns=['gender', 'name'])
movie_crew_df.reset_index(inplace=True)
movie_crew_df.rename(columns={"id": "crew_member_id"}, inplace=True)
movie_crew_df.rename(columns={"index": "id"}, inplace=True)
movie_crew_df = movie_crew_df[['id', 'movie_id', 'crew_member_id', 'job', 'department', 'credit_id']]
movie_crew_df['id'] = movie_crew_df['id'].apply(lambda x: x + 1)
movie_crew_df
```

### Crew members

```python
crew_members_df = crew_df[['id', 'name', 'gender']].copy()
crew_members_df.drop_duplicates(subset=['id'], inplace=True)
crew_members_df.sort_values('id', inplace=True)
crew_members_df
```

### Departments

```python
crew_dept = crew_df['department']

unique_dept = []
for dept in crew_dept:
    if dept not in unique_dept:
        unique_dept.append(dept)
        
crew_department_df = pd.DataFrame(data=unique_dept, columns=["department"])
crew_department_df
```

### Jobs

```python
crew_jobs = crew_df['job']

unique_jobs = []
for job in crew_jobs:
    if job not in unique_jobs:
        unique_jobs.append(job)
        
crew_jobs_df = pd.DataFrame(data=unique_jobs, columns=["job"])
crew_jobs_df
```

```python
jobs = crew_df[['job', 'department']].copy()
grouped_jobs = jobs.groupby('department')

unique_grouped_jobs = []
for key, item in grouped_jobs:
    unique_grouped_jobs.append(list(item['job'].unique()))

departments = list(grouped_jobs.groups.keys())
```

```python
rows = []
for dep_index, dep in enumerate(unique_grouped_jobs):
    for job in dep:
        rows.append({ "job": job, "department": departments[dep_index] })

jobs_df = pd.DataFrame(rows, columns=['job', 'department'])
jobs_df.reset_index(inplace=True)
jobs_df.rename(columns={"index": "id"}, inplace=True)
jobs_df['id'] = jobs_df['id'].apply(lambda x: x + 1)
jobs_df
```

```python
jobs_ix = []
for crew in tqdm(movie_crew_df.values):
    jobs_ix.append(jobs_df[(jobs_df['job'] == crew[3]) & (jobs_df['department'] == crew[4])]['id'].values[0])

movie_crew_df['job_id'] = jobs_ix
movie_crew_df.drop(columns={"job", "department"}, inplace=True)
movie_crew_df = movie_crew_df[['id', 'movie_id', 'crew_member_id', 'job_id', 'credit_id']]
movie_crew_df
```
